```python
import csv
import json
import os
```

```python
scenic_scenarios_file_path = "/home/dferna3/llm_reasoning/scenic_scenarios"
json_file = "../data/scenarios.json"
output_csv = "../data/results.csv"
latency_output_csv = "../data/latency_results.csv"
```

## For a scenarios.json file

```python
try: 
    # Load the JSON file
    with open(json_file, 'r') as file:
        data = json.load(file)

    # Ensure 'Scenarios' is in the JSON structure
    if 'Scenarios' not in data or not isinstance(data['Scenarios'], list):
        print("The JSON file does not have a valid 'Scenarios' array.")
    else: 
        # Extract scenario IDs and count the scenarios
        scenarios = data['Scenarios']
        scenario_count = len(scenarios)

        # Create/replace the results.csv file
        with open(output_csv, 'w', newline='') as csvfile:
            csvwriter = csv.writer(csvfile)

            # Write the header
            csvwriter.writerow(['scenario_id', "ground_truth"])

            # Write each scenario ID
            for i, scenario in enumerate(scenarios, start=1):
                ground_truth = scenario.get("ground_truth", "")  # Default to empty string if not present
                csvwriter.writerow([i, ground_truth])

        print(f"File '{output_csv}' created/replaced successfully with {scenario_count} scenarios.")

except FileNotFoundError:
    print(f"The file '{json_file}' was not found.")
except json.JSONDecodeError:
    print("The JSON file is not valid.")
except Exception as e:
    print(f"An error occurred: {e}")
```
