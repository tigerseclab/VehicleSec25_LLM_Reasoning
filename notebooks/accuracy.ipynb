```python
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
# Load the CSV file
file_path = "../data/Accuracy.csv"
df = pd.read_csv(file_path)
```

```python
# Display the first few rows to understand the structure (optional)
print(df.head())

# The second column is ground_truth
ground_truth = df.iloc[:, 1]

# Dictionary to store accuracies
accuracies = {}

# Loop through each model's results starting from the third column
for model in df.columns[2:]:
    predictions = df[model]
    accuracy = accuracy_score(ground_truth, predictions)
    accuracies[model] = accuracy

# Convert accuracies to DataFrame for better visualization
accuracy_df = pd.DataFrame(list(accuracies.items()), columns=['Model', 'Accuracy'])
# Sort the DataFrame by 'Accuracy' in descending order
accuracy_df_sorted = accuracy_df.sort_values(by='Accuracy', ascending=False).reset_index(drop=True)
accuracy_df_sorted

```
